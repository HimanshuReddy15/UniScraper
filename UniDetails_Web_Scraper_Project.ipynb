```python
#import libraries
from bs4 import BeautifulSoup
import requests
import datetime
import time
import smtplib
```

```python
#Connect to the Website
URL = 'https://yocket.com/universities/university-of-texas-arlington-824'

Headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36"}

page = requests.get(URL,headers=Headers)

soup1 = BeautifulSoup(page.content,'html.parser')

soup2 = BeautifulSoup(soup1.prettify(),'html.parser')

Uni = soup2.find(class_="text-yocket-neutral-900 cursor-default").text.strip()
if 'at ' in Uni:
    [Name,Place] = Uni.split('at ')
elif ', ' in Uni:
    [Name,Place] = Uni.split(', ')
else:
    [Name,Place] = [Uni,'-']

Location = soup2.find(class_="text-sm md:text-base font-medium text-yocket-neutral-500").text.strip()
[State,Country] = Location.split(', ')

Details = soup2.find_all("p",class_="text-base font-bold text-yocket-neutral-900")
Data = [data.text.strip() for data in Details[:3]]
Courses = [data.text.strip() for data in Details[3:]]
address_element = soup2.find(class_="text-yocket-neutral-900 text-sm font-medium md:max-w-[14.875rem]")
if address_element:
    Address = address_element.text.strip()
else:
    Address = "Info Not Found"
try:
    email_element = soup2.find_all(class_="text-yocket-neutral-500 text-sm font-medium")[2]
    Email = email_element.text.strip()
except IndexError:
    Email = "Info Not Found"


print(Name)
print(Place)
print(State)
print(Country)
print(Data)
print(Courses)
print(Address)
print(Email)
```
